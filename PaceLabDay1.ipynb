# NUMPY

```python
import numpy as np

# Create an array
a = np.array([[1,2],[3,4]])
print("Array a: ")
print(a)

#Shape and dtype
print("\nDimention of matrixx :",a.shape)
print("Data type of a :", a.dtype)
```

```python
# Basic Matrix Operation

print("Matrix Multiplication\n",a*2)
print("\nMatrix Square\n",a**2)
print("\nMatrix Transpose\n", a.T)
print("\nMatrix Product\n",a@(a.T))
```

```python
# Useful functions

print("Zeroes: \n",np.zeros((2,3)))
print("Ones: \n",np.ones((3,4)))
print("Range: \n",np.arange(0,10,2))
print("Linespace: \n",np.linspace(0,10,20))
"Trignometry: \n",np.cos(0)
```

# PANDAS

```python
import pandas as pd

#Creating a DataFrame
data = {
    'Name' : ['Joel','Bob','Charlie'],
    'Age' : [19,22,35],
    'Score' : [96,87.4,45.3]
}
df = pd.DataFrame(data)
print("DataFrame: \n")
print(df)
print("\n")

#Accessing data from df
print(df.head(2)) # access firt 2 row of data
print("\n",df.tail(2)) # access last 2 row of data
print("\n",df.describe()) # display all the central tendencies
print("\n",df['Name']) #print the specific columns from the table
```

# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Sine
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
#Cos
x = np.linspace(0,20,100)
y = np.cos(x)

plt.plot(x,y)
plt.title("Cosin Wave")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show()
```

```python
# Bar Chart

categories = ['A','B','C','D']
values = [23,75,46,68]
plt.bar(categories,values,color = ['Orange','Yellow','Green','Red']) # If we give 2 colour then it alternates
plt.title("Bar Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Tutorial 1

n = int(input("Enter the number of students"))
names=[]
cgpas=[]
for i in range(n):
  name = input("Enter the name:")
  cgpa = float(input("Enter the cgpa:"))
  names.append(name)
  cgpas.append(cgpa)
plt.bar(names,cgpas,color='Orange') # If we give 2 colour then it alternates
plt.title("Bar Graph")
plt.xlabel("Name of Student")
plt.ylabel("CGPA")
plt.show()
```

```python
# Histogram

data = np.random.randn(1000) # generate 1000 numbers which has mean 0 and standard deviation 1
plt.hist(data,bins=30,color='Green',edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
